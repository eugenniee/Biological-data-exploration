```python
euk = pd.read_csv("eukaryotes.tsv ", sep = "\t")
my_types = {
    "Species": "string",
    "Kingdom": "string",
    "Class": "string",
    "Assembly status": "string",
    "Number of genes": "Int64",
    "Number of proteins": "Int64",
}

euk = pd.read_csv("eukaryotes.tsv", sep = "\t", dtype = my_types, na_values = ["-"])
```

```python
euk["Size (Mb)"] > 500
```

```python
mask = euk["Size (Mb)"] > 500
mask
```

```python
# Get a new dataframe with the rows for which the comparison was true.

euk[mask]
```

```python
euk[euk["Size (Mb)"] > 500]
```

```python
# Find genomes belonging to a single species.

euk[euk["Species"] == "Penicillium expansum"]
```

```python
# Find all genomes that are not fish

euk[euk["Class"] != "Fishes"]
```

```python
# Find bird and fish genomes

euk[euk["Class"].isin(["Birds", "Fish"])]
```

```python
# Find genomes for species starting with the letter Q
euk[euk[("Species")].str.startswith("Q")]
```

```python
# Making complex conditions
# FInd genomes between 100 and 1000 Mb

(euk["Size (Mb)"] > 100) & (euk["Size (Mb)"] < 1000)
```

```python
euk[(euk["Size (Mb)"] > 100) 
    & (euk["Size (Mb)"] < 1000)]
```

```python
# When we want to find values between two limits

euk[euk["Size (Mb)"].between(100, 1000)]
```

```python
# Find genomes that are not birds or fishes, or have low GC
# ~ means NOT

euk[~(euk["Class"].isin(["Birds", "Fishes"])) | (euk["GC%"] < 40)]
```

```python
# Filtering on columns with missing data

(euk["Number of genes"] > 10000).value_counts(dropna=False)
```

```python
# If we want to include genomes with an unknown number of genes
# Using fillna to replace missing data with a value bigger than cutoff

len(euk[euk["Number of genes"].fillna(20000) > 10000])
```

```python
# Size of all human genomes

euk[euk["Species"] == "Homo sapiens"]["Size (Mb)"]
```

```python
# What is the mean genome size for birds?

(
    euk[euk["Class"] == "Birds"] #find bird genomes
    ["Size (Mb)"] #get the size column
    .mean()
)
```

```python
# What is the median number of genes for genomes with GC < 50?

(
    euk[euk["GC%"] < 50]
    ["Number of genes"]
    .median()
)
```

```python
# How many genomes were sequenced for each kingdom in 2010?

euk[euk["Publication year"] == 2010]["Kingdom"].value_counts()
```

```python
# The largest number of proteins for fish genomes from 2017

(
    euk[
        (euk["Class"] == "Fishes") & (euk["Publication year"] == 2017)
    ] 
    ["Number of proteins"]
    .max()
)
```

```python
idx=(euk["Class"] == "Fishes") & (euk["Publication year"] == 2017)
euk.loc[idx, 'Number of proteins'].max()
```

```python
# If the data is messy: capitals and lower cases in "Class" column

euk[~(euk["Class"].str.lower().isin(["birds", "fishes"])) | (euk["GC%"] < 40)]
```
