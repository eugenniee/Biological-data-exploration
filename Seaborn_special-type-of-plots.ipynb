```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

path_to_file = "C:/Users/Evgeniia/Desktop/Work/Data-Analysis/example_files/contigs.csv"
con = pd.read_csv(path_to_file)
con.head()
```

```python
con.dtypes
```

```python
len(con)
```

```python
# Visualize relationship between GC content and length

g=sns.relplot(data=con, x="GC", y="length")
plt.title("GC content vs length", pad=10)
```

```python
# Filtering for outliers

g=sns.relplot(data=con[con["length"] < 100000], x="GC", y="length")
plt.title("GC content vs length", pad=10)
```

```python
g=sns.relplot(data=con[con["length"] < 100000], x="GC", y="length", alpha=0.01)
plt.title("GC content vs length", pad=10)

# alpha makes the plot transparent; between 0 and 1
```

```python
# change size of the points (s) and remove the border around each point (linewidth)

sns.relplot(data=con[con["length"] < 100000], x="GC", y="length", s=1, linewidth=0)
```

```python
# plot a sample

sns.relplot(data=con[con["length"] < 100000].sample(1000), x="GC", y="length")
```

```python
# When the data is scattered unevenly, it's difficult to capture all data in a scatterplot. Trying hexbin
# The color of each cell above tells us how many points fall into the region. 
# The marginal histograms show us the distribution for each axis separately.

g = sns.jointplot(
    data=con[con["length"] < 100000], x="GC", y="length",
    kind="hex", marginal_ticks=True,
    gridsize=30  # the number of hexagons in the x-direction and the y-direction
)

# Setting a supertitle. We have to store the result of the plot, use the fig attribute to get a reference to the figure.

g.fig.suptitle("GC content vs length", y=1.05)
```

```python
g = sns.jointplot(
    data=con[con["length"] < 40000], x="GC", y="length",
    kind="hex", marginal_ticks=True,
    gridsize=30  # the number of hexagons in the x-direction and the y-direction
)

# Setting a supertitle. We have to store the result of the plot, use the fig attribute to get a reference to the figure.

g.fig.suptitle("GC content vs length", y=1.05)
```

```python
# Drawing a contour plot

g = sns.jointplot(
    data=con[con["length"] < 40000], x="GC", y="length",
    kind="kde",
    shade=False, color="red",
    n_levels=10  # number of contours
)

g.fig.suptitle("GC content vs length")
```

```python
sns.jointplot(
    data=con[con["length"] < 40000], x="GC", y="length",
    kind="kde",
    shade=True, color="red",
    n_levels=20,
    cmap="OrRd"
)
```
