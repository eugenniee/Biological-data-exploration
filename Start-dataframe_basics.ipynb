```python
# Specify that several columns (Species, Kingdom, Class, Assembly) should be stored as strings. 

my_types = {
    "Species": "string",
    "Kingdom": "string",
    "Class": "string",
    "Assembly status": "string",
}

euk = pd.read_csv("eukaryotes.tsv", sep = "\t", dtype = my_types)
euk.dtypes
```

```python
# How to tell pandas that the dash symbol means missing data.

euk = pd.read_csv("eukaryotes.tsv", sep = "\t", dtype = my_types, na_values = ["-"])
euk.tail()
```

```python
euk.dtypes
```

```python
# Pandas assumes that any numerical column with missing data must be floating point (like "Number of genes", "Number of proteins").
# We're adding these columns to the dtype dictionary.

my_types = {
    "Species": "string",
    "Kingdom": "string",
    "Class": "string",
    "Assembly status": "string",
    "Number of genes": "Int64",
    "Number of proteins": "Int64",
}

euk = pd.read_csv("eukaryotes.tsv", sep = "\t", dtype = my_types, na_values = ["-"])
euk.dtypes
```

```python
# How to get more information about the data frame.
# No missing values in the "Publication year" column, that's why "int64" has a lowercase i.

euk.info()
```

```python
# How to change column names.

my_types = {
    "Species": "string", "Kingdom": "string", "Class": "string", "Assembly status": "string", "Number of genes": "Int64", "Number of proteins": "Int64",
}

my_names = [
    "species", "kingdom", "class", "size", "gc", "genes", "proteins", "year", "status"
]

pd.read_csv(
    "eukaryotes.tsv",
    sep = "\t",
    dtype = my_types,
    na_values = ["-"],
    #header = 0,
    names = my_names,
).head()
```

```python
# How to change column names, v2.

my_types = {
    "Species": "string", "Kingdom": "string", "Class": "string", "Assembly status": "string", "Number of genes": "Int64", "Number of proteins": "Int64",
}

my_names = [
    "species", "kingdom", "class", "size", "gc", "genes", "proteins", "year", "status"
]

df = pd.read_csv(
    "eukaryotes.tsv",
    sep = "\t",
    dtype = my_types,
    na_values = ["-"],
)

df.columns = my_names
df.head()
```

```python
df = pd.read_csv(
    "eukaryotes.tsv",
    sep = "\t",
    dtype = {"Species": "string", "Kingdom": "string", "Class": "string", "Assembly status": "string", "Number of genes": "Int64", "Number of proteins": "Int64"},
    na_values = ["-"],
)

df = df.rename(columns={"Species": "species", "Kingdom": "kingdom", "Class": "class", "Size (Mb)": "size"})

# Or rename the existing DataFrame (rather than creating a copy)
# df.rename(columns={"Species": "species", "Kingdom": "kingdom", "Class": "class", "Size (Mb)": "size"}, inplace=True)

df.head()
```
